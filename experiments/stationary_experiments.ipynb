## Plotting

```python
def heatmap(reg: str, ax: Axes, cbar_ax: Axes | None = None, out_of_dist: bool = False):
    variables = globals()
    sns.heatmap(
        variables[("out_" if out_of_dist else "") + "reg_" + reg.lower()],
        vmin=0, # lower bound of cmap
        vmax=1, # upper bound of cmap
        annot=True,
        cmap="viridis",
        linewidth=0.5, # spacing between squares
        xticklabels=[l if len(l) <= 5 else l[:4]+"." for l in labels],
        yticklabels=[l if len(l) <= 5 else l[:4]+"." for l in labels],
        ax=ax,
        cbar_ax=cbar_ax
    )
    ax.set_title(f"Reg. {reg.upper()}")
    for _, spine in ax.spines.items():
        spine.set_visible(True) # borders around heatmap
```

```python
fig, axs = plt.subplots(2, 5, figsize=(12, 5))
cbar_ax = fig.add_axes((1, .05, .03, .95))
fig.tight_layout(pad=1.5)
heatmap("A", axs[0,0], cbar_ax)
heatmap("B", axs[0,1], cbar_ax)
heatmap("C", axs[0,2], cbar_ax)
heatmap("D", axs[0,3], cbar_ax)
heatmap("E", axs[0,4], cbar_ax)
heatmap("F", axs[1,0], cbar_ax)
heatmap("G", axs[1,1], cbar_ax)
heatmap("H", axs[1,2], cbar_ax)
heatmap("I", axs[1,3], cbar_ax)
heatmap("J", axs[1,4], cbar_ax)
plt.show()
```

```python
vs = globals()
rs = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
ls = [(1,0),(2,0),(1,2),(2,1)]
n=1000
p=0.5
a="greater"
pre="out_reg_"

plt.style.use('default')
fig, axs = plt.subplots(2, 1, figsize=(7, 6))
cbar_ax = fig.add_axes((1, .05, .03, .95))
fig.tight_layout(pad=2.5)

sns.heatmap(
    np.array([[binomtest(int(vs[pre[4:]+r][*l]*n),n,p,a).pvalue for r in rs] for l in ls]).round(2),
    vmin=0, # lower bound of cmap
    vmax=1, # upper bound of cmap
    annot=True,
    cmap="bone",
    linewidth=0.5, # spacing between squares
    xticklabels=[r.upper() for r in rs],
    yticklabels=[(labels[l1] + " $\\leq$ " + labels[l2]) for l1, l2 in ls],
    ax=axs[0],
    cbar_ax=cbar_ax
)
axs[0].set_title("In-distribution")
for _, spine in axs[0].spines.items():
    spine.set_visible(True) # borders around heatmap
sns.heatmap(
    np.array([[binomtest(int(vs[pre+r][*l]*n),n,p,a).pvalue if (pre+r) in vs.keys() else np.nan for r in rs] for l in ls]).round(2),
    vmin=0, # lower bound of cmap
    vmax=1, # upper bound of cmap
    annot=True,
    cmap="bone",
    linewidth=0.5, # spacing between squares
    xticklabels=[r.upper() for r in rs],
    yticklabels=[(labels[l1] + " $\\leq$ " + labels[l2]) for l1, l2 in ls],
    ax=axs[1],
    cbar_ax=cbar_ax
)
axs[1].set_title("Out-of-distribution")
for _, spine in axs[1].spines.items():
    spine.set_visible(True) # borders around heatmap
```
