## Plotting

```python
def heatmap(reg: str, ax: Axes, cbar_ax: Axes | None = None, out_of_dist: bool = False):
    variables = globals()
    prefix = ("out_" if out_of_dist else "") + "reg_" + reg.lower()
    lab = variables[(prefix[4:] if out_of_dist else prefix) + "_labels"]
    sns.heatmap(
        variables[prefix + "_mean"][::variables[prefix + "_mean"].shape[0]-1, ::variables[prefix + "_mean"].shape[1]-1],
        vmin=0, # lower bound of cmap
        vmax=1, # upper bound of cmap
        annot=True,
        cmap="viridis",
        linewidth=0.5, # spacing between squares
        xticklabels=lab[::len(lab)-1],
        yticklabels=lab[::len(lab)-1],
        ax=ax,
        cbar_ax=cbar_ax
    )
    ax.set_title(f"Reg. {reg.upper()} (σ={round(variables[prefix + '_std'], 3)})")
    for _, spine in ax.spines.items():
        spine.set_visible(True) # borders around heatmap
```

```python
fig, axs = plt.subplots(2, 5, figsize=(12, 5))
cbar_ax = fig.add_axes((1, .05, .03, .95))
fig.tight_layout()
axs[0,0].axis('off') # gap for reg A
heatmap("B", axs[0,1], cbar_ax)
heatmap("C", axs[0,2], cbar_ax)
heatmap("D", axs[0,3], cbar_ax)
heatmap("E", axs[0,4], cbar_ax)
heatmap("F", axs[1,0], cbar_ax)
heatmap("G", axs[1,1], cbar_ax)
heatmap("H", axs[1,2], cbar_ax)
heatmap("I", axs[1,3], cbar_ax)
heatmap("J", axs[1,4], cbar_ax)
plt.show()
```

```python
out_of_dist = False

regs_full = ["B", "C", "D", "E", "F", "G", "H", "I", "J"]
regs_out = ["C", "D", "F", "G", "H", "I"]
regs = regs_out if out_of_dist else regs_full

variables = globals()
means = []
stds = []

for reg in regs:
    prefix = ("out_" if out_of_dist else "") + "reg_" + reg.lower()
    means.append(variables[prefix + "_mean"][-1,0])
    stds.append(variables[prefix + '_std'])
means = np.array(means)
stds = np.array(stds)

print(np.array([binomtest(k=int(result*1000), n=1000, p=0.5, alternative="greater").pvalue for result in means]).round(2))

plt.style.use('seaborn-v0_8')
plt.rcParams.update({'lines.markeredgewidth': 1})

plt.bar(regs, means)
plt.errorbar(regs, means, yerr=stds, fmt='none', ecolor=(0, 0, 0, 0.75), capsize=5.0)
plt.axhline(y=0.5, linestyle='--', linewidth=1.0, color=(0, 0, 0, 0.5))
plt.ylim(0, 1)
plt.xlabel("Regulation")
plt.ylabel("Mean Win Rate")
plt.show()
```
