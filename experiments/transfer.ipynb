## Plotting

```python
def heatmap(reg: str, ax: Axes, cbar_ax: Axes | None = None, out_of_dist: bool = False):
    variables = globals()
    prefix = ("out_" if out_of_dist else "") + "reg_" + reg.lower()
    sns.heatmap(
        variables[prefix + "_mean"],
        vmin=0, # lower bound of cmap
        vmax=1, # upper bound of cmap
        annot=True,
        cmap="viridis",
        linewidth=0.5, # spacing between squares
        xticklabels=variables[(prefix[4:] if out_of_dist else prefix) + "_labels"],
        yticklabels=variables[(prefix[4:] if out_of_dist else prefix) + "_labels"],
        ax=ax,
        cbar_ax=cbar_ax
    )
    ax.set_title(f"Reg {reg.upper()} (σ={round(variables[prefix + '_std'], 3)})")
    for _, spine in ax.spines.items():
        spine.set_visible(True) # borders around heatmap
```

```python
fig, axs = plt.subplots(2, 3, figsize=(8, 5))
fig.suptitle("Win rate of row player after regular transfer (3 seeds; n=1000)")
cbar_ax = fig.add_axes((1, .05, .03, .9))
fig.tight_layout()
heatmap("B", axs[0,0], cbar_ax)
heatmap("D", axs[0,1], cbar_ax)
heatmap("E", axs[0,2], cbar_ax)
heatmap("G", axs[1,0], cbar_ax)
heatmap("H", axs[1,1], cbar_ax)
heatmap("J", axs[1,2], cbar_ax)
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
fig.suptitle("Win rate of row player after regular transfer (3 seeds; n=1000)")
cbar_ax = fig.add_axes((1, .05, .03, .9))
fig.tight_layout()
heatmap("C", axs[0], cbar_ax)
heatmap("F", axs[1], cbar_ax)
heatmap("I", axs[2], cbar_ax)
plt.show()
```
